```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data1.csv')
data
```

What analysis task we can perform on data
→Total no of notification sent
→Total clicks
→Click through rate
→Click by notification type
→Click by device type
→Descriptive Stats
→Descriptive stats by category



→Total no of notification sent

```python
Total_no_of_notification_sent = data.count()
Total_no_of_notification_sent
```

```python
Total_no_of_clicks = data['Clicked'].sum()
Total_no_of_clicks
```

→Click by notification type

```python
Click_by_notification_type = data.groupby('Notification_type')['Clicked'].sum()
Click_by_notification_type
```

→Click by device type

```python
Click_by_device_type = data.groupby("Device_Type")["Clicked"].sum()
Click_by_device_type
```

→Descriptive Stats

```python
Descriptive_status = data["Clicked"].describe()
Descriptive_status
```

visualzation work plotting a bargraph 

```python
Click_by_notification_type.plot(kind='bar',color='yellow')
plt.title('Clicks by Notification Type')
plt.xlabel('Notification Type')
plt.ylabel('Number of Clicks')
plt.grid()
```

```python
Click_by_device_type.plot(kind='bar',color='red')
plt.title('Clicks by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Number of Clicks')
plt.grid()
```
